```python
import matplotlib.pyplot as plt
import numpy as np
import re

%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
def get_data(filename):
    loss, acc = [], [] # assumes (train, val) dims
    with open(filename) as fi:
        for ln in fi:
            if ln.startswith('train loss'):
                ln = ln.rstrip()
                loss.append(re.findall(r'loss: (\d+\.\d+)', ln))
                acc.append(re.findall(r'acc: (\d+\.\d+)', ln))

    loss = np.stack(loss).astype(float)
    acc = np.stack(acc).astype(float)
    return loss, acc
```

```python
# loss01, acc01 = get_data('../logs/e01.out')
loss02, acc02 = get_data('../logs/seed/ours_029h4s1.out')
```

```python
# plt.plot(loss01[:,0], label='RAM', ls='-', color='red')
# plt.plot(loss01[:,1], ls='--', color='red')

plt.plot(acc02[:,0], label='train', ls='-', color='black')
plt.plot(acc02[:,1], label='val', ls='-', color='red')
# plt.ylim(80,100)
plt.legend()
plt.grid(alpha=0.2)
```
