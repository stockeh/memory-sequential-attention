```python
import os
import numpy as np
import matplotlib.pyplot as plt

VOL = '/s/farley/a/nobackup/stock/data/cs/cluttered-mnist'
```

```python
dim = 60
mnist_cluttered = 'mnist_cluttered_60x60_6distortions.npz'

data = np.load(os.path.join(VOL, mnist_cluttered))
Xtrain, Ttrain = data['x_train'], np.argmax(data['y_train'], axis=-1).reshape(-1,1)
Xval, Tval = data['x_valid'], np.argmax(data['y_valid'], axis=-1).reshape(-1,1)
Xtest, Ttest = data['x_test'], np.argmax(data['y_test'], axis=-1).reshape(-1,1)

Xtrain = Xtrain.reshape((Xtrain.shape[0], 1, dim, dim))
Xval = Xval.reshape((Xval.shape[0], 1, dim, dim))
Xtest = Xtest.reshape((Xtest.shape[0], 1, dim, dim))

print("Train:", Xtrain.shape, Ttrain.shape)
print("Val", Xval.shape, Tval.shape)
print("Test", Xtest.shape, Ttest.shape)

print(Xtrain.min(), Xtrain.max())
print(np.unique(Ttrain, return_counts=True))
```

```python
X, T = Xtrain, Ttrain

fig, axs = plt.subplots(2,5, figsize=(10,4))
axs = axs.ravel()

for i,c in enumerate(np.unique(T)):
    axs[i].imshow(X[T==c][0], cmap='gray_r')
    axs[i].set_xticks([], []), axs[i].set_yticks([], [])
    
fig.tight_layout()
```
